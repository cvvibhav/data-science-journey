**Today, we tackle a problem called the "Curse of Dimensionality."**

Day 25: We had 1 column of data (Years of Experience). Easy to plot on a line.

Day 26: We had 3 columns (Size, Rooms, Age). Harder, but we can imagine 3D.

***Today: What if we have 30 columns? Or 1,000 columns (like pixels in an image)?**

We can't draw a 1,000-dimension graph. Our brains explode. We need a way to crush those 30 columns down to just 2 columns, without losing the important information.



# ***`We use a tool called PCA (Principal Component Analysis).`***

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# 1. Load the Data

data = load_breast_cancer()
X = data.data
y = data.target

print(f"Original shape : {X.shape}")
```

# 2: Scale the Data

```python
# 2. Standardize (Make everything Scale 0 to 1-ish)
# PCA is sensitive to size. If one column is "Salary" (50,000) and another is "Age" (30), PCA gets confused. We must shrink them to be the same scale.
scaler = StandardScaler()

X_scaled= scaler.fit_transform(X)
```

# Run PCA

```python
#create PCA model

pca = PCA(n_components = 2)

# model fit and transform
x_pca = pca.fit_transform(X_scaled)

print(f"New Shape : {x_pca.shape}")
```

# Visualize the Result

```python
# 5. Plot the 2 columns
# X_pca[:, 0] is the 1st Principal Component (x-axis)
# X_pca[:, 1] is the 2nd Principal Component (y-axis)
# c=y means "Color it by the label" (Cancer vs Healthy)


plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=y, cmap='plasma')

plt.title("30 dimentions shrinked to 2 dimentions")
plt.xlabel('First Principle component')
plt.ylabel('Second Principle component')
plt.show()
```
