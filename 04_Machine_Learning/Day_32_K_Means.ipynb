**K-Means Clustering.**

Understand Unsupervised Learning and use K-Means to automatically group similar data points into clusters.

# 1. Setup


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# create fake data 
# we know there are four groups but the robot doesnt know

X,y_trui = make_blobs(n_samples = 300, centers = 4 , cluster_std = 0.60 , random_state = 0)
```

```python
# 2. Visualize the Mess
# Notice we just plot X. We don't use 'y' (color) because the robot is "blind" to the answers.

plt.scatter(X[:,0],X[:,1], s=50 , color = 'gray' )
plt.title('The Raw Data (Unlabled)')
plt.show()
```

# 2: Train K-Means

```python
# 3 . create the model

kmeans = KMeans(n_clusters = 4)

# 4. fit the model 
# Notice: We only pass X! No 'y' allowed!

kmeans.fit(X)

#5. results

y_means = kmeans.predict(X)
```

# 3: Visualize

```python
# Plot the dots, colored by the robot's guess

plt.scatter(X[:,0],X[:,1], c=y_means , s = 50 , cmap= 'viridis')

# Plot the Centers (The "Means") - Black dots
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1] ,c ='black', s = 200 , alpha = 0.5)

plt.title("K-Means Clustering Results")
plt.show()
```

```python
inertia = []

# try k from 1 to 50 
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# plot 

plt.plot(range(1,11), inertia, marker = 'o')
plt.title('the elbow method')
plt.xlabel('number of clusters')
plt.ylabel('Inertia(error)')
plt.show()
```
