# Yesterday, we used ARIMA. It is mathematically pure, but it is "dumb" about the real world.

It doesn't know that sales spike on **Christmas.**

It doesn't know that traffic drops on **Saturdays.**

It crashes if data is missing.

Today, we use Prophet. Developed by **Facebook (Meta)**, this tool was built because their own internal analysts were too slow at tuning *ARIMA models.* They built an ***"Analyst in a Box."***

It is designed for Business Data: it handles holidays, weekends, and missing data automatically.

## Use Prophet to forecast time series data with strong seasonal patterns and holidays, and visualize the "Components" (Trend, Weekly, Yearly).

You don't need to tune (p, d, q). You just feed it data, and it figures out the cycles.

### CRITICAL RULE: 

**Prophet requires exactly two columns with specific names:**

***ds: The Date (Datestamp).***

***y: The Value (Metric).***

**If you name them "Date" and "Sales", it will crash.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

#create fake data
#lets simulate a store that is busy on weekends

dates = pd.date_range(start = '2024-01-01' , periods = 365 , freq = 'D')
np.random.seed(42)
```

```python
#trend + seasonality + random
# we make weekends (day 5 and 6) have higher values

values = np.linspace(10,50,365)+ np.random.normal(0,2,365)
values += [10 if d.dayofweek >=5 else 0 for d in dates]
```

```python
#strict fromatting for prophet

df = pd.DataFrame ({'ds':dates , 'y' : values})

print(df.head())
```

# Train the Model

```python
#.fit

# we ismplt instnantiate and fit

model = Prophet()
model.fit(df)
```


# The Future Dataframe


```python
# Prophet needs a blank template to fill in.

#create fiture dates

future = model.make_future_dataframe(periods = 30)

#predcit
forecast = model.predict(future)

print(forecast[['ds','yhat' , 'yhat_lower', 'yhat_upper']].tail())
```

# The "Components" Plot

```python
model.plot(forecast)
plt.title('Prophet Forecast')
plt.show()
```

## What the plot shows:

**Black dots**  historical data

**Blue line**  forecast (yhat)

**Shaded area**  uncertainty range

**Right side**  future predictions

**You should see:**

***An upward trend***

***`Repeating weekly waves (weekend spikes)`***

```python
# Visualize the Components (Trend, Weekly, Yearly)

model.plot_components(forecast)
plt.show()
```

# Look at the Components Plot (the second set of graphs).

### Trend: You should see a straight line going up. (The business is growing).

### Weekly: You should see a massive spike on Saturday/Sunday and a drop on weekdays.

### Value: This tells you exactly how much "Weekend Effect" you have.

### Yearly: (Since we only simulated one year and didn't add strong yearly seasonality, this might look flat or random, which is expected).

**Why is this useful? If your boss asks: "Why were sales high yesterday?"**

ARIMA says: "Because yesterday's number was 52.4." (Useless).

***`Prophet says: "Because it was a Saturday (+10 effect) and the overall Trend is up (+2 effect)." (Actionable).`***
