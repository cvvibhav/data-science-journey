Yesterday, we used the Naive Forecast (Yesterday's Price). It was accurate, but "jittery." It reacted to every single tiny bump in the road.

Today, we learn to ignore the bumps and see the road. We are using Moving Averages (Rolling Windows).

Instead of looking at just one day, we look at the last 7 days (or 30 days) and take the average. This filters out the noise and reveals the true Trend.

# ***`Understand Smoothing using .rolling() to calculate Simple Moving Averages (SMA) and visualize long-term trends against daily noise. Key Skill: Rolling Windows & Trend Identification.`***

The Concept: The Dog Walker 
Imagine a person walking a dog on a long leash.

The Dog (Daily Price): Runs back and forth, sniffing bushes, chasing squirrels. Its path is jagged and messy.

The Human (Moving Average): Walks in a straight, smooth line down the sidewalk.

If you want to know where they are going next:

Don't look at the dog (Noise).

Look at the human (Trend).

The Math:

Day 7 SMA: Average of Day 1, 2, 3, 4, 5, 6, 7.

Day 8 SMA: Average of Day 2, 3, 4, 5, 6, 7, 8. (The window "rolls" forward).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#recreate data of fake stocks

dates = pd.date_range(start='2024-01-01' , end='2024-12-31', freq='D')
np.random.seed(42)
price_changes = np.random.uniform(-2,2, size=len(dates))
prices = 100 + np.cumsum(price_changes)

df = pd.DataFrame({'Date': dates , 'Price': prices})
df.set_index('Date', inplace = True)
```

```python
print(df)
```

# 2: Calculate The Moving Average

```python
# 2. Rolling Window Calculation
# .rolling(window=7) looks at the last 7 rows
# .mean() calculates the average of those 7 rows

df['SMA_7']  = df['Price'].rolling(window = 7).mean()
df['SMA_30']  = df['Price'].rolling(window = 30).mean()


# Check the first 10 rows
# Note: The first 6 rows of SMA_7 will be NaN (Not enough data yet)

print(df.head(10))
```

# 3: Visualize the Smoothness

```python
plt.figure(figsize= (12,6))

#plot the raw price

plt.plot(df.index, df['Price'] , label = "Daily Price", color = 'gray', alpha = 0.3)

#plot the trends
plt.plot(df.index, df['SMA_7'] , label = "7 Day Average", color = 'orange', linewidth = 2)
plt.plot(df.index, df['SMA_30'] , label = "30 Day Average", color = 'orange', linewidth = 3)


plt.title('Stock price vs Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
