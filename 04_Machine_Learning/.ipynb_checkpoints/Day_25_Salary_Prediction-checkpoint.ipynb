 # 1: Setup & Data Creation

Since we don't have a CSV file handy, we will generate a realistic one using Python.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# 1. Create a Data
# random years of experiecnce (0 to 10 years , 30 employees)

X = np.random.rand(30,1)*10

# base salary 30k + 8k every year and some bonuses
y = 3000+(X*8000) + np.random.randn(30,1)*4000

# lets peek at the first five people
print("Here is our imaginary data : ")
print(X[:5])
```

# 2: Cut the Deck (Train-Test Split) We cut the data into two piles.

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size = 0.2 , random_state = 50)

print(f"studying with {len(X_train)} people.")
print(f"Testing on {len(X_test)} people.")
```

# 3: Teach the Robot We only show the Robot the Training pile.

```python
model = LinearRegression()
#to learn from study guide
model.fit(X_train , y_train)
```

# 4: The Final Exam Now we bring out the Secret Test pile (X_test) and ask the robot to guess.

```python
# "Robot, guess the salary for these new people!"
predictions = model.predict(X_test)
```

```python
# Let's see how close it got
print("Real Salaries vs Robot Guess:")

for i in range(5):
    print(f"Real : {y_test[i][0]:.0f} | Guess : {predictions[i][0]:.0f}")
```

# 5: Draw the Picture Let's draw a line through the dots to see the Robot's rule.

```python
# Show the Training dots (Blue)
plt.scatter(X_train , y_train , color= 'blue')

#draw the red line (the rule the robot learned)
plt.plot(X_train , model.predict(X_train), color = 'red')

plt.title('Salary vs years of experience')
plt.xlabel('years')
plt.ylabel('Salary ($)')
plt.show()
```

```python
print("Slope (raise per extra year):", model.coef_)
```
