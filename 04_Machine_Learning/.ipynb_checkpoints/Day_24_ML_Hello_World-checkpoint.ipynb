# Traditional Programming vs. Machine Learning

### Traditional: You give the computer the Rules (Code) and the Data. It gives you the Answers.

    Analogy: You give a recipe and ingredients to a chef. They give you a cake.

### Machine Learning: You give the computer the Data and the Answers. It figures out the Rules.

    Analogy: You give the chef a finished cake and the ingredients, and say "Figure out the recipe."

# Supervised vs. Unsupervised Learning

### Supervised Learning (Teaching with Flashcards):

    You show the computer a picture of a cat and say "This is a cat." You show a dog and say "This is a dog."

    Eventually, you show a new picture, and it guesses.

    Key Terms: Features (Input data, X) and Target (The Answer, y).

### Unsupervised Learning (Organizing a Messy Room):

    You give the computer a pile of photos with no labels. You just say "Group similar things together."

    It puts all the sharp-eared animals in one pile and floppy-eared ones in another.

# Scikit-Learn (sklearn)

This is the Pandas of ML. It is the industry-standard library in Python for building models.

**1. Setup**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#import sklearn
from sklearn.linear_model import LinearRegression
```

**2. The "Hello World" of ML**

Let's teach a computer to double numbers (y = 2x).

But we won't write def double(x): return x * 2. We will force the machine to learn that rule.

```python
# 1. Prepare Data (Features must be 2D , Target is 1D)

# X= Inputs(Questions)
X = np.array ([[1],[2],[3],[4],[5]])

# Y = Targets(Answers)
y = np.array([2,4,6,8,10])

#2. Initialize the model 
model = LinearRegression()

# 3. Train the Model 
# 'Hera the Ques are X and y are answers , FIND THE PATTERN
model.fit(X,y)

# 4. predict
#asking the machine to predict the number it has never seen (10)
prediction = model.predict([[10]])

print("Prediction for input 10 is :")
print(prediction[0])
```
