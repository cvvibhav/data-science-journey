# ***`Understand Time Series Decomposition (Trend + Seasonality + Residuals) and use the seasonal_decompose tool to break a dataset apart.`***

# Equation Data = Trend + Seasonality + Noise

```python
# fake dataset


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
dates = pd.date_range(start= '2024-01-01',periods=730, freq = 'D')

#a trend, prices go up evert day (0.1,0.2,0.3.....)
trend = np.linspace(0,50,730)

#seasonality , a wave repeasts every 365 days
seasonality = 10*np.sin(np.linspace(0,3.14*4,730))

#noise

noise= np.random.normal(0,2,730)

# mix em all
total_sales = trend + seasonality + noise

#create df
df = pd.DataFrame({'Date': dates , 'Sales': total_sales})
df.set_index("Date", inplace = True)


#plot


plt.plot(df.index,df['Sales'])
plt.title("The Raw Data (Trend + Seasonality + Noise)")
plt.show()
```

# 2: Decompose It

```python
# perform decomposition
# model='additive' means Data = Trend + Season + Noise

result = seasonal_decompose(df['Sales'] , model = 'addictive', period = 365)

#plot

result.plot()
plt.show()
```

# Look at the 4 graphs generated by result.plot():

Observed: The messy original chart (The Mix).

Trend: A perfectly straight line going up. It found our "Ingredient A"!

Seasonal: A perfect wavy line repeating every year. It found our "Ingredient B"!

Resid (Residual): A fuzzy mess of static. This is the Noise we added.

Why this matters: If you are analysing MBA sales data:

Use Trend to tell investors: "Our company is growing 10% year-over-year."

Use Seasonality to tell Supply Chain: "Stock up on inventory every November."
