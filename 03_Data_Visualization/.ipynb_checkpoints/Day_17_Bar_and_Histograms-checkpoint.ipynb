# Bar Chart (plt.bar):

Purpose: Compares categorical data.

Structure: The X-axis has names (e.g., "Apple", "Banana", "Cherry"). The Y-axis has values (counts or averages).

Visual: The bars usually have gaps between them because the categories are distinct.



 # Histogram (plt.hist):

Purpose: Shows the distribution of numerical data.

Structure: The X-axis has number ranges (called bins). The Y-axis tells you how many items fall into that range.

Visual: The bars touch each other because the number line is continuous.

Example: If you have a list of 100 people's ages, a histogram shows how many are 0-10, 11-20, 21-30, etc.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Create a Bar Chart

```python
# plt.bar(categories, values)

#data 
lang = ['R','Python','C++','Java']
users = [60,100,45,90]

#plot bat
plt.bar(lang,users, color='skyblue') #change the color to 'purple' or '#FF5733' (hex code).

#labels
plt.title("Programming Popularity")
plt.xlabel("language")
plt.ylabel("Users (millions)")

plt.show()
```

# Create a Histogram

np.random.normal(mean, std_dev, size) generates random numbers from a normal (Gaussian) distribution.

30 → the mean (average) of the distribution

5 → the standard deviation

200 → how many random values to generate

✔️ What the result looks like

ages will be a NumPy array containing 200 simulated ages that are:

centered around 30 years old

with most values typically within ± 5 years (so ~25–35)

but with some random variation, as a normal distribution allows

✔️ Why use this?

This is commonly used for:

simulating realistic datasets

testing models

generating synthetic demographic data

```python
#generate 100 random ages between 18 and 60 
#np.random.normal(mean,std_dev, count)

ages = np.random.normal(30,5,200)

#plot Histogram
# 'bins' determines how many buckets to split the data into
plt.hist(ages, bins=20 , color = 'orange' , edgecolor = 'black')

#labels
plt.title("Age Distribution of participants")
plt.xlabel('age')
plt.ylabel('count of ppl')

plt.show()
```
