# ***ðŸ“˜ 1. KDE (Kernel Density Estimate) â€” "The Silk Sheet"***

The Problem

A histogram divides data into binsâ€”like building a staircase with Lego blocks.
The edges are sharp, and sometimes you lose the true shape of the data.

The KDE Solution

A KDE plot is like draping a smooth silk sheet over that Lego staircase.
It follows the general shape but removes the rough blockiness.

Why Use KDE?

1. Shows the flow and smooth trend of your data
2. Removes noise caused by arbitrary bin sizes
3. Great for comparing distributions between groups

# ***ðŸ“˜ 2. Correlation (r) â€” "The Accelerator & The Brake"***

Correlation is a number between -1 and +1 that tells you how two variables move together.

**ðŸ”º +1.0 Positive Correlation â€” Accelerator**

Press accelerator â†’ car speed goes up.
Both variables increase together.

Example:

Height vs. Weight (usually positive)

**ðŸ”» -1.0 Negative Correlation â€” Brake**

Press brake â†’ car speed drops.
One increases, the other decreases.

Example:

Outside temperature vs. number of winter jackets sold

**âšª 0.0 No Correlation â€” Radio Volume**

Turning car radio louder doesnâ€™t change the speed.
The variables have no meaningful relationship.

# ***ðŸ“˜ 3. Heatmap â€” "Thermal Vision"***

The Problem

A correlation table (like 10Ã—10 or 50Ã—50) contains too many numbers.
Reading them one by one is slow and painful.

The Heatmap Solution

A heatmap is like switching to thermal vision goggles:

1. Red (or Dark Red) â†’ Strong Positive Relationship (Accelerator)
2. .Blue (or Dark Blue) â†’ Strong Negative Relationship (Brake)
3. White / Light Color â†’ No relationship (Radio Volume)

You donâ€™t read numbersâ€”you see patterns.

# Setup & Import

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#load iris dataset

df = sns.load_dataset('iris')
print(df.head())
```

# Advanced Distributions (KDE)

lets look at sepal_length , and use kde = True to see the smooth line

```python
#1. Histogram with KDE
plt.figure(figsize=(8,6))
sns.histplot(data = df , x = 'sepal_length', kde = 'True', color = 'purple', bins = 20)
plt.title("SEPAL LENGTH DISTRIBUTION")
plt.show()
```

```python
# Compare Species: Use hue to see the separation
# this proves that different species have different physical properties
```

```python
sns.histplot(data=df , x = 'sepal_length', hue = 'species' ,element = 'step')
plt.show()
```

```python
sns.histplot(data=df , x = 'sepal_width', hue = 'species' ,element = 'step')
plt.show()
```

```python
sns.histplot(data=df , x = 'petal_length', hue = 'species' ,element = 'step')
plt.show()
```

```python
sns.histplot(data=df , x = 'petal_width', hue = 'species' ,element = 'step')
plt.show()
```

# The Pairplot (The "Bird's Eye View")

The Pairplot (The "Bird's Eye View")

```python
sns.pairplot(df, hue='species', palette ='husl' )
plt.show()
```

***Observation:*** Look for the charts where the colors are completely separated. That means those features are really good at predicting the species!

# The "Thermal Vision" (Heatmap)

```python
# 1. Calc the 'correlation matrix'
# numeric only = true ignores the species name

corr_matrix = df.corr(numeric_only = True)

#2. Put on Thermal Goggles

plt.figure(figsize= (8,6))

# annot = true writes the number in boxes
#cmap = 'coolwarm' makes +1 red and -1 blue

sns.heatmap(corr_matrix , annot = True , cmap = 'coolwarm' , fmt = '.2f')

plt.title("Correlation Matric : RED == Accelerator and Blue == Brake")
plt.show()
```

# **1. Themes (style)**
Think of this as the "Instagram Filter" for your entire notebook. It changes the background color and grid lines.

You set this once at the top of your notebook: sns.set_theme(style="...")

**"darkgrid" (Default):** Grey background with white grid lines. Good for seeing values clearly.

**"whitegrid" (Professional):** White background with grey grid lines. Best for presentations/reports.

**"dark":** Grey background, no grid.

**"white":** White background, no grid. Very clean.

**"ticks":** White background, but adds little tick marks on the axes.

## **2. Palettes (palette)**
This controls the colors of the data. Choosing the wrong palette can make your chart misleading.

***There are 3 main types of palettes:***

**A. Qualitative (For Categories)**

Use these when the groups are distinct and have no order (e.g., Male vs Female, Smoker vs Non-Smoker).

Names: 'deep' (default), 'muted', 'pastel', 'bright', 'dark', 'colorblind'.

Example: sns.countplot(x='sex', palette='pastel')

**B. Sequential (For Ranges)**

Use these when data goes from Low to High (e.g., Age, Fare, Total Bill). The color gets darker as the number gets higher.

Names: 'Blues', 'Greens', 'Reds', 'Oranges', 'Purples', 'Greys'.

Special Modern Ones: 'rocket' (dark red/orange), 'mako' (teal/blue), 'viridis'.

**C. Diverging (For Extremes)**

Use these when there is a middle point (zero) and you want to show positive vs. negative (e.g., Correlation Heatmaps).

Names: 'coolwarm' (Blue to Red), 'vlag' (Blue to Grey to Red), 'icefire'.

Example: sns.heatmap(corr, cmap='coolwarm') (Note: Heatmaps use cmap instead of palette, but it's the same lists).

#  Understanding x, y, and hue (Seaborn)  
This guide explains the basic concepts of **x**, **y**, and **hue** and how they are used in **1D, 2D, and 3D visualizations**.

---

##  What is `hue`?
`hue` adds **color-based grouping** to your plot.  
It splits your data into different colored subgroups.

Example:  
- Without hue â†’ all points/bars are the same color  
- With hue â†’ each category gets a different color  
  (e.g., species, gender, categories, petal size)

---

##  1D Plot (One Dimension + Hue)
A **1D plot** shows the distribution of **one numeric variable**.  
We use `x` (or sometimes `y`) and add `hue` to compare groups.

### Example meaning:
- `x = sepal_length` â†’ numeric variable  
- `hue = species` â†’ show each species in a different color  

This lets us compare distributions using color.

---

##  2D Plot (Two Dimensions + Hue)
A **2D plot** has **two numeric axes**:
- `x` â†’ horizontal axis  
- `y` â†’ vertical axis  
- `hue` â†’ color groups  

### Example meaning:
- `x = sepal_length`  
- `y = sepal_width`  
- `hue = species`  
This shows how flower species differ in 2 variables at the same time.

---

##  3D Plot (Two Dimensions + Hue + Size/Style/Facet)
Seaborn doesn't make true 3D,  
so we add a **third dimension** using:
- `size=`  
- `style=`  
- or faceting (`col=` or `row=`)

This gives a **3D-like visualization**.

### Example meaning:
- `x = sepal_length`  
- `y = sepal_width`  
- `hue = species`  
- `size = petal_length` â†’ third dimension (bubble size)

This helps show relationships among **three variables at once**.

---

## Summary Table

| Concept | Meaning | What It Controls |
|--------|---------|------------------|
| **x** | Horizontal axis variable | Categories or numeric values |
| **y** | Vertical axis variable | What you're measuring |
| **hue** | Color grouping | Split data by groups |
| **1D** | One numeric variable | Distribution |
| **2D** | Two numeric variables | Scatter or bar relationships |
| **3D** | Two axes + hue + size/style | Deeper multi-variable insight |

---



```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris, x='sepal_length', hue='species', fill=True)
plt.title("1D Plot: Sepal Length Distribution (Split by Species)")
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width',
                hue='species', palette='coolwarm')
plt.title("2D Plot: Sepal Length vs Sepal Width (Split by Species)")
plt.show()
```

```python
plt.figure(figsize=(25, 20))
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species',          # categorical dimension
    size='petal_length',    # 3rd dimension (numeric)
    sizes=(20, 200),
    palette='coolwarm'
)
plt.title("3D Plot: Sepal Length vs Width (Hue: Species, Size: Petal Length)\n")
plt.show()
```
