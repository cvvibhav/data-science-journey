# List and Loops

```python
my_list = [1, 2, 3]   #list
for number in my_list:  #for loop
    # This code block is INDENTED
    print(number * 2)

# Output would be:
# 2
# 4
# 6
```

```python
#our list of tasks

todo_list = ["Start Day 4","Practise Python", "Go for a walk"]
print(f"Original List : {todo_list}")

#1 add a new item at the end 
todo_list.append("Journal My day")
print(f"after Append : {todo_list}")

#2 remove the item by its name
todo_list.remove("Go for a walk")
print(f"after removing [go for a walk] from the list : {todo_list}")

#3 remove the last item 
finished_task = todo_list.pop()
print(f"Finished task: {finished_task}")
print(f"Final list: {todo_list}")
```

```python
tools = ["Python", "Git", "Jupyter", "Pandas", "Scikit-learn"]

print("tools in learning")
for tool in tools :
    print(f"- {tool}")

print("\n LOOP IS FINISHED")
```

# Task 4: Mini-Challenge (Loop + Method) (1 hour)

Goal: Create a new list of numbers that are the double of an original list.

This is a very common pattern in data science.

```python
numbers = [1,5,10,20]
doubled_numbers =[] #start with an empty list

#loop over original list
for num in numbers:
    doubled_num = num*2   #calc the double
    doubled_numbers.append(doubled_num)  #addend the doubled number in to the empty list

print(f"Original : {numbers}")
print(f"Doubled  : {doubled_numbers}")
```
