```python
contacts = [{"name": "Mahesh", "Phone":"8988989889"}, {"name":"Raju","Phone":"7899877890"}]
```

```python
#create this  in datavase [list]
```

```python
contacts = []
# Function 1: Add a contact
# This uses dictionaries and the .append() method

def add_contact(name,phone):
    new_contact = {'name': name,'phone': phone}
    contacts.append(new_contact)
    print(f"Success : Added {name} to your contacts")
```

```python
# Function 2: View all contacts
# This uses a 'for' loop

def view_contacts():
    print("----- YOUR CONTACTS -----")
    if not contacts:
        print("Your Contact Book is empty.")
        return
        
    for contact in contacts:
        name = contact.get("name","Unknown Name")
        phone = contact.get("phone","No Phone")
        
        print(f"Name: {name}, Phone: {phone}")
    print("---------------")
```

```python
# Function 3: Find a contact by name

def find_contact(name_to_find):
    for contact in contacts:
        #use "if" to check the name
        if contact["name"] == name_to_find:
            print(f"Match Found : Name: {contact['name']}, Phone: {contact['phone']}")
            return   #return stops the function once we find a match
# This 'else' is part of the 'for' loop (a bit advanced, but cool!)
# It runs only if the loop finishes without a 'return'

    print(f"Sorry we could'nt find the name : {name_to_find}")
```

```python
# Add some contacts
add_contact("Alice", "123-456")
add_contact("Bob", "789-012")
add_contact("Charlie", "555-555")

#view all contacts
view_contacts()

#find specific contact
find_contact("Bob")

#finding a contact that doesnt exist
find_contact("Vibhav")
```
