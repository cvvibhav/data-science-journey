# IF / ELSE

```python
age = 19
if age>=18:
     print(f"access granted as you are {age} ")
else:
    print("Access Denied")
```

```python
#lets go if/elif/else (Grade finder)
```

```python
score = int(input("What is your Score"))

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 55:
    print("Grade: D")
elif score >= 35:
    print("Grade: E")
else:
    print("Grade: F")
```

# First Function (def)

```python
# 1. DEFINE the function (the recipe)

def greet():
    print("Hello! Welcome to Jupyter.")
    print("we are learning about funtions")

# 2. CALL the function (run the recipe)\
greet()
```

```python
# Let's make our greeting personal.
```

```python
def greet_person(name):
    print(f"Hello {name}, WELCOMEE...")

#calling fn 
greet_person("Mani")
greet_person("Deepak")
```

```python
def add_numbers(num1 , num2):
    total =  num1 + num2
    return total

#call fn and store its output in var
sum_ = add_numbers(5,10)
print(f"The Result is : {sum_}")
```
