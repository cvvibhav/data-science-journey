 # **What is a Dictionary? and how does it differ from Lists ?**
 
Think of a list as a filing cabinet with numbered drawers (0, 1, 2...). A dictionary is a filing cabinet with labeled drawers ("name", "age", "city"). It's an unordered collection of item 

# *List vs. Dictionary*

List: Ordered, accessed by numerical index. my_list[0]

Dictionary: Unordered, accessed by a unique key. my_dict["name"]

Syntax: Key-Value Pairs Dictionaries use curly braces {}. Each item is a "pair" consisting of a key (the label) and a value (the data).

```python
# The key is "name" , the value is "Vibhav"
# The key is "age" , the value is 30

my_profile = {"name" : "Vibhav", "age":30,"is_learning":True}

# my_profile["age"] = 22 (Modifies the existing value)
# my_profile["email"] = "vibhav@example.com" (Adds a new key-value pair)
my_profile["name"] = "Vibhav C V"
my_profile["age"] = 22
my_profile["email"]= "vibhav@example.com" #new key value pair added
print(my_profile)
```

```python
print(my_profile['name'])
```

```python
my_profile={ "name":"Vibhav", "Skills": ["Python","Git","Jupyter"]}  #A list inside a dictionary

#get the second skill
print(my_profile["Skills"][1])  #would print "GIT"
```

# Dictionary

```python
my_car = {
    "make": "Toyota",
    "model": "Camry",
    "year": 2021,
    "features": ["Bluetooth", "Backup Camera", "Sunroof"]
}

print(my_car)
```

```python
#print model
print(my_car["model"])

#print the first feature of car
print(my_car["features"][0])
```

# Accessing & Modifying Data

```python
# 1. Change the year
# 2. Add a new key-value pair for color
# 3. See the whole dictionary

my_car["make"] = 2022
print(my_car["make"])  #1
print() #line gap

my_car["color"] = "white"
print(my_car["color"])  #2
print()

print(my_car)  #3
```

# Looping Over a Dictionary

You can't just loop over a dictionary; 

***you need to use the .items() method to get both the key and value.***

```python
print("------Car Details--------")
for key, value in my_car.items():
    print(f"Key : {key}|| Value : {value}")
```
