```python
import pandas as pd
import numpy as np 
```

| Feature   | Series                 | DataFrame                  |
| --------- | ---------------------- | -------------------------- |
| Structure | 1D (one column)        | 2D (rows + columns)        |
| Contains  | Values + index         | Multiple Series            |
| Columns   | No                     | Yes                        |
| Shape     | (n,)                   | (n, m)                     |
| Example   | `s = pd.Series([...])` | `df = pd.DataFrame({...})` |


# Creating a Series (1D)

```python
# create a series from a list

s= pd.Series([10,20,30,40], index=['a','b','c','d'])
print(f"Pandas series:\n{s}")
```

```python
#access by label

print(f"value at label 'b' is : {s['b']}")
```

# Creating a DataFrame (2D)

```python
# Creating a Dictionary 

data= {'name': ['raju','bheem','chutki','jaggu'], 'age': [22,34,29,7], 'City':['dholakpur','mirajpur','rampur','agnipur']}

#creating a dataframe

df_from_dict = pd.DataFrame(data)

print(f"A dataframe from a dict :\n {df_from_dict}")
```
