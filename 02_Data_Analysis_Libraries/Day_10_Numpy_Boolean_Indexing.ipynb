You get a new array of True/False values. We call this a Boolean array or a "mask".

CRITICAL: You must use & and |, not the Python words and or or. You also must use parentheses () around each condition.

```python
# np.where() #(The "if/else" for Arrays): This is the NumPy version of an if/else statement. It's incredibly useful.
```

```python
scores = np.array([77,82,95,60,91,53,88])
```

```python
#1. Create a "mask"

passing_mark = scores>70
print(f"Mask: {passing_mark}")
```

```python
#2. Apply the mask to filter the data

passing_score = scores[passing_mark]
print(f"Passing Scores : {passing_score}")
```

```python
#3. Do it in one line (common way)

a_grade = scores[scores >=90]
print(f"A grades : {a_grade}")
```

```python
# Get the "B Grades" (between 80 and 89)

b_grade = scores[(scores >=80) & (scores<89)]
print(f"B Grade : {b_grade}")
```

```python
# Get the "At Risk" scores (less than 60)

risk_rate = scores[(scores<=65)]
print(f"Risk Scores : {risk_rate}")
```

```python
#np.where()

#label all scores as "Pass" or "Fail"

pass_fail = np.where(scores>= 60, "Pass","Fail")
print(f"P/F : {pass_fail}")
```
