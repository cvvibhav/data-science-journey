Learn to aggregate data using summary statistics (.describe(), .mean(), .value_counts()) and the powerful .groupby() method.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
print(df.head())
```

```python
# 1. Get the 8-number summary for all numerical columns

print("--- .describe() ---")
print(df.describe())
```

```python
# 2. Get a single statistic for a single column

avg_sepal_length = df['sepal_length'].mean()
print(f"Average Sepal Length is :{avg_sepal_length}")

avg_sepal_width = df['sepal_width'].mean()
print(f"Average Sepal width is :{avg_sepal_width}")

avg_petal_length = df['petal_length'].mean()
print(f"Average Petal Length is :{avg_petal_length}")

avg_petal_width = df['petal_width'].mean()
print(f"Average Petal Width is :{avg_petal_width}")
```

```python
# 3. Get the sum of a column

total_petal_length = df['petal_length'].sum()
print(total_petal_length)
```

# .value_counts() (Categorical Data)

```python
# This is one of the most-used functions

species_counts = df['species'].value_counts()
print("---.value_counts()---")
print(species_counts)
```

```python
sepl_counts = df['sepal_length'].value_counts()
print("---.value_counts()---")
print(sepl_counts)
```

# The .groupby() Method

```python
# 1. Group by 'species' and then apply the .mean() function

species_avg = df.groupby('species').mean()
print("\n==============Avg Measurements per Species==================")
print(species_avg)
```

```python
# 2. Group by 'species' and get the max

spc_max = df.groupby('species').max()
print(f"The Max [Species] : \n{spc_max}")
```
