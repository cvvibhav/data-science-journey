What is Matplotlib? It is the most established plotting library in Python. Almost every other library (like Seaborn or Pandas plotting) is built on top of Matplotlib.

The "Pyplot" Interface: We rarely use the whole library. We use a specific module called pyplot, which provides a simple interface like MATLAB.

Standard Import: import matplotlib.pyplot as plt

The Anatomy of a Plot:

Figure: The whole window or page that everything is drawn on.

Axes: The actual graph (the x and y axis area) where data is plotted.

Axis: The number lines (x-axis and y-axis).

Basic Workflow:

plt.plot(x, y): Draws the data.

plt.title("..."): Adds a title.

plt.xlabel("..."): Labels the x-axis.

plt.show(): The final command that displays the plot to your screen.

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import numpy as np # We need this to generate fake data

# This special command ensures plots show up INSIDE the notebook
%matplotlib inline
```

```python
#1. Create Data

days = ['Mon','Tue','Wed','Thur','Fri']
study = [2,3,1.5, 4,2.5]

#plot data

plt.plot(days, study)

#3. Display the Plot

plt.show()
```

Styling Your PLot

```python
# Color: 'g' (green), Marker: 'o' (dots), Line style: '--' (dashed)

plt.plot(days,study, color = 'green', marker = 'o', linestyle = 'dashed')

# To add labels

plt.title("Study Hours")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.show()
```

```python
days = ['Mon','Tue','Wed','Thur','Fri']
study = [2,3,1.5, 4,2.5]
sleep = [6,6.5,7,7,6.5]

# plot line 1

plt.plot(days,study, label= 'Study', color = 'blue', marker = 'o', linestyle= 'dotted')

# plot line 2

plt.plot(days,sleep, label= 'Sleep', color = 'orange', marker = 'x', linestyle= 'dashed')

# Add Title & Labels
plt.title("Study vs Sleep Habits")
plt.xlabel("Day")
plt.ylabel("Hours")

# Add a Legend (Critical when you have 2+ lines
plt.legend()
plt.show()
```
