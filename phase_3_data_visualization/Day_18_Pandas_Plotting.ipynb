Master Pandas Plotting: Using the df.plot() method to visualize data directly from a DataFrame without extracting lists.

```python
# Instead of writing plt.plot(df['col_a'], df['col_b']), Pandas lets you write: df.plot(x='col_a', y='col_b', kind='line')
```

The kind Parameter: This is the most important switch. It tells Pandas what type of chart to draw.

kind='line' (Default): Line chart.

kind='bar': Bar chart (Vertical).

kind='barh': Bar chart (Horizontal) - Great for long labels!

kind='hist': Histogram.

kind='scatter': Scatter Plot (New today!). (+ve corelation , -ve corelation , no corelation )

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Setup & Load Data

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
print(df.head())
```

# The Scatter Plot (Relationships)

```python
# kind='scatter' requires an 'x' and a 'y' column
#Let's see if longer sepals also mean wider sepals.

df.plot(kind='scatter', x = 'sepal_length', y = 'sepal_width', title = 'Sepal Length vs Width')
plt.show()
```

# The Histogram (Distributions)

```python
# Let's look at the distribution of petal_length.
```

```python
# kind='hist' only needs one column (y)
# 'bins' works just like in matplotlib

df['petal_length'].plot(kind = 'hist', bins = 20 , title = 'Petal Length Distribution')
plt.show()
```

# The Bar Chart (Aggregated Data)

Crucial Step: You usually cannot plot raw data as a bar chart. You must Aggregate it first (using value_counts or groupby).

Goal: Plot the average petal length per species.

```python
#Step 1 :  aggregate the data
#group by species and getthe mean 

df_grp = df.groupby('species')['petal_width'].mean()
print(df_grp)
```

```python
# Step 2: Plot the aggregated data
# we use kind = 'bar'
df_grp.plot(kind = 'bar', color = ['red','green', 'blue'] , title = 'Average Petal Length per species' )
plt.ylabel('Length(cm)')
plt.show()
```
