# Bill Splitter

```python
total_bill = 124.5
num_friends = 4

# calc & print the cost per person

cost_per_person = total_bill / num_friends
print(f"the cost per person comes down to {cost_per_person} rupees")
```

# Comparison 

```python
my_age = 30
friend_age = 25

#lets check with boolean fn (comparison) to know if its true or false
print(my_age > friend_age)  # Is my age greater than theirs?
print(my_age == friend_age)  # Are our ages the same?
print(my_age != friend_age)  # Are our ages the different ?

#checking the data type

comparison = (my_age > friend_age)
print(type(comparison))
```

# Lists

```python
# List of Data Science Tools
tools = ["Python", "Git","Jupyter","Pandas"]

# List of Numbers
primes= [2,3,5,7,11,13]

# Mixed List that can hold anything
my_profile = ["Vibhav" , 21 , "INDIAN" , True]

print(tools)
print(primes)
print(my_profile)
```

```python
#get the FIRST item

first_tool = tools[0]
print(f"my first tool is : {first_tool}.")
```

```python
# Get the THIRD item .

third_tool = tools[2]
print(f"my third tool is : {third_tool}")
```

```python
# Get the LAST item using negative indexing

last_tool = tools[-1]
print(f"Last item is : {last_tool}")
```
